### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as stats
````

````python
# Generate an array of 100 elements following a normal distribution
mu, sigma = 0, 0.1 # mean and std
normal_dist = np.random.normal(mu, sigma, 100)
count, bins, ignored = plt.hist(normal_dist, 30, density=True)
plt.plot(bins, 1/(sigma * np.sqrt(2 * np.pi)) *
               np.exp( - (bins - mu)**2 / (2 * sigma**2) ),
         linewidth=2, color='r')
plt.show()
normal_dist
````

````python
# Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom
chi_square = np.random.chisquare(3,size=100)
count, bins, _ = plt.hist(chi_square, 30, density = True)
x = np.linspace(0, 20, 100)
plt.plot(x, stats.chi2.pdf(x, df=3))
plt.xlim([0, 20])
plt.show()
chi_square
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import math
data = [4, 2, 5, 8, 6]
data = np.array(data)
# Numpy
# Esta es la mejor opción, la otra es poco óptima. ddof = 1. 
# Para indicar que es la desviación estándar muestral. N - ddof
st_dev = data.std(ddof=1)
print("Standard deviation of the given data: " + str(st_dev))
# Formula
print("List : " + str(data))
mean = sum(data) / len(data)
def variance():
    num = 0
    for i in data:
        num += (i - mean)**2
    return num/(len(data)-1)
st_dev = math.sqrt(variance())
print("Standard deviation of the given data (Formula): " + str(st_dev))
````

````python
# Python
import math
def mean_calc(data):
    value = 0
    for i in data:
        value += i
    value=value/len(data)
    return value

def std_calc(data):
    sd = 0.0
    m = mean_calc(data)
    n = len(data)
    for i in data:
        sd += (i-m)**2

    sd = math.sqrt(sd/(n-1))

    return sd
std_calc(data)
````
